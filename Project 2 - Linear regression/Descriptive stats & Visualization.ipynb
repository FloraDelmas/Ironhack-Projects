```python
import pandas as pd
```

```python
df=pd.read_csv('HR-Employee-Attrition.csv')
```

```python
df.head()
```

```python
df.describe().T
```

```python
drop_columns=['HourlyRate','DailyRate','EmployeeCount','EmployeeNumber','StandardHours','MonthlyRate','Over18']
df=df.drop(drop_columns,axis=1)
```

```python
df.columns
```

```python
df.head()
```

#### Create categories

```python
Distance_Home=['1','2','3','4']
cutoffs = [0,7,14,22,29]
df['Distance_Home']=pd.cut(df['DistanceFromHome'],cutoffs, labels=Distance_Home)
```

```python
df.Distance_Home=df.Distance_Home.astype(int)
```

```python
Cat_Ages=['18-30','31-40','41-50','51-60']
cutoffs = [17,30,40,50,61]
df['Cat_Age']=pd.cut(df['Age'],cutoffs, labels=Cat_Ages)
```

```python
df.OverTime.value_counts()
```

```python
df.dtypes
```

#### Transform my target column in dummies

```python
df['Attrition_bol']=0
df.loc[(df.Attrition=='No'),['Attrition_bol']]=0
df.loc[(df.Attrition=='Yes'),['Attrition_bol']]=1
```

```python
column_order=[ 'Gender','Age','Cat_Age' ,'Attrition','Attrition_bol', 'Department','JobRole','JobInvolvement',
              'JobLevel',  'JobSatisfaction',
              'BusinessTravel', 'DistanceFromHome','Distance_Home','MonthlyIncome','PercentSalaryHike', 'PerformanceRating',
              'OverTime','Education', 'EducationField', 'EnvironmentSatisfaction', 'MaritalStatus', 'RelationshipSatisfaction',
              'NumCompaniesWorked',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
       'YearsSinceLastPromotion', 'YearsWithCurrManager']
df=df[column_order]
```

```python
df=df.drop('DistanceFromHome', axis=1)
```

```python
df.to_csv('EmployeeAttritionv1.csv')
```

```python
sns.set(style="white")
corr = df.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 15))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 20, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

```python
fig = plt.gcf()
fig.set_size_inches(16, 10)
sns.set(font_scale=1.4)

sns.scatterplot(x='MonthlyIncome', y='TotalWorkingYears', hue='Attrition',
                data=df, s=45, palette=["red", "royalblue"])

plt.ylabel('Years')
plt.title('Relation between Monthly Income and Total working years')
```
